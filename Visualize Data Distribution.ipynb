ต่อไปเป็นการสร้าง list เปล่าให้ colors #เป็นการกำหนดค่าที่ดูตาหลักที่4

ต่อไปเป็นการกำหนดสีแต่ละตัว โดยกำหนดค่าและกำหนดสี

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
        
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c=colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=colors)   #30 คือขยายขนาด
```

ต่อไปคือการเขียนแบบไม่ใส่ colors เข้าไป โดยใช้ c เป็นตัวที่ 4เลย วิธีการก็คือเป็นการกำหนด colors map

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap='YlOrBr')   #เพิ่ม ,c=setosa_np[:,3],cmap='YlOrBr' เพื่อกำหนด C
```

ต่อไปคือการแทนค่าสีด้วย rgb แต่และ แม็คเกอร์
