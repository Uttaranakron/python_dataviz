### use boxplot to compare features between 3 classes 
ให้เปรียบเทียบแต่ละ Features ของดอกไม้ทั้ง 3 Classes

```python
import numpy as np
```

```python
data1 =  [setosa_np[:,0],versicolor_np[:,0],virginica_np[:,0]]
```

```python
data1
```

```python
matplotlib.rcParams['figure.figsize'] = [15,15]
plt.subplot(1,4,1)
data1 = [setosa_np[:,0],versicolor_np[:,0],virginica_np[:,0]]
plt.boxplot(data1,labels = ['Setosa','Versicolor','Virginica'])
plt.title('Feature 1 Comparison') ;
plt.subplot(1,4,2)
data2 = [setosa_np[:,1],versicolor_np[:,1],virginica_np[:,1]]
plt.boxplot(data2,labels = ['Setosa','Versicolor','Virginica'])
plt.title('Feature 2 Comparison') ;
plt.subplot(1,4,3)
data3 = [setosa_np[:,2],versicolor_np[:,2],virginica_np[:,2]]
plt.boxplot(data3,labels = ['Setosa','Versicolor','Virginica'])
plt.title('Feature 3 Comparison') ;
plt.subplot(1,4,4)
data4 = [setosa_np[:,3],versicolor_np[:,3],virginica_np[:,3]]
plt.boxplot(data4,labels = ['Setosa','Versicolor','Virginica'])
plt.title('Feature 4 Comparison') ;
```

## Bubble Chart
#### คือการใช้ขนาดของวงกลมแสดงค่าของ Data

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=1000*setosa_np[:,2],c='r',alpha = 0.5)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=1000*setosa_np[:,2],cmap='cool', alpha=0.4, edgecolors="pink", linewidth=2,marker = 'D')
```

```python

plt.scatter(virginica_np[:,3],virginica_np[:,0],s=2000*virginica_np[:,1], c = virginica_np[:,2], cmap="Blues",alpha=0.8, edgecolors="grey", linewidth=2)
plt.xlabel("virginica is Features 4") ##ชื่อแกน X
plt.ylabel("Virginica is Features 1") ##ชื่อแกนY
plt.title("Virginica bubble plot") ##ชื่อ Bubble chart
plt.show()
```

```python

map.choropleth(geo_data=country_geo,data=data_to_plot,
                          columns=['Region and province',2549],
                          key_on='feature.properties.name',
                          fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
                          legend_name='Average Debt')
```
