## 1> สร้าง function ชื่อ smartDict  ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ที่สองเป็น list ของ text 
- ถ้า list แรกเป็นเลขคู่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า list แรกเป็นเลขคี่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น
เช่น

output1 = smartDict( [ 1 , 2 ,3 ] , [ 'apple' , 'banana' , 'orange'] , True )

จะได้ output1 เป็น { 1:'apple' , 'banana' : 2, 3 :'orange' }



```python
def smartDict(listtext, listnumber, logicV):
    if logicV : #logicv == True
        retdict = {} #สร้างdictเปล่าขึ้นมา
        for index in range(len(listtext)):  
            retdict[listtext[index]]= listnumber[index] #dictทีละตัว
    else:
        retdict = {} 
        for index in range(len(listtext)):  
             retdict[listnumber[index]]= listtext[index] 
                    
            
    return retdict  #ส่งค่าoutput
```

```python
smartDict( [ 1 , 2 ,3 ] , [ 'apple' , 'banana' , 'orange'] , True )
```

```python
smartDict( [ 1 , 2 ,3 ] , [ 'apple' , 'banana' , 'orange'] , False)
```

#### คำถามที่1 - กดอมิโนใด (A, T, G, C) มีปริมาณมากที่สุดและมีเท่าไร

```python
DNAlong = 0 
for row in range(DNAseq.shape[0]):
    DNAlong = DNAlong + len(DNAseq.iloc[row][0]) #DNAlong += en(DNAseq.iloc[row][0]
    
print(f'sequence is {DNAlong} charaters long')
```

```python
def countChar(seq,character):
    charcount = 0
    for char in seq:
         if char == character:
                charcount = charcount + 1
    return charcount
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGA','A')
```

```python
ANum = 0 
TNum = 0
GNum = 0 
CNum = 0 
for row in range(DNAseq.shape[0]):
    ANum += countChar(DNAseq.iloc[row][0],'A') 
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
    
print(f'There are {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs')
```

กรดอะมิโนที่มีปริมาณมากที่สุดคือ T และมีค่ามากที่สุดคือ 9594

#### คำถามที่ 2 - จงหา sequence ที่มีขนาด 3  ที่เกิดขึ้นซ้ำมากที่สุด

```python
def populate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-4):
        str_list.append(seq[i:i+3])
    return str_list
```

```python
populate3charstring('ATTAAAGGTTT')
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]):
    list_all_3char = list_all_3char + populate3charstring(DNAseq.iloc[row][0]) 
```

```python
len(list_all_3char)
```

```python
unique_3char = list(set(list_all_3char))
```

```python
len(unique_3char)
```

```python
dict_3char = {}
for sub_str in unique_3char:
    num = 0
    for s in list_all_3char:
        if sub_str == s:
            num += 1
    dict_3char[sub_str] = num
```

```python
dict_3char 
```

```python
maxVal = -100
maxKey = None
for k in dict_3char.keys():
    if dict_3char[k] > maxVal:
        maxVal = dict_3char[k]
        maxKey = k
```

```python
maxKey
```

```python
print(f'max 3-char-sequence is {maxKey} with frequency equals {maxKey}')
```
